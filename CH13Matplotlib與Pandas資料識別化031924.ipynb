## 繪製簡單的折線圖(Ch13_1_1.py)

```python
# 目的是使用 Matplotlib 套件來繪製一個折線圖。

# Line18 說明，import matplotlib.pyplot as plt
# 此行導入 Matplotlib 中的 pyplot 模組，並使用 plt 作為別名，以便在後續的程式碼中調用它的功能。

# Line20 說明，data = [-1, -4.3, 15, 21, 31]
# 這行建立一個名為 data 的 Python 列表(list)，其中包含要在折線圖中繪製的數據點。

# Line21 說明，plt.plot(data)
# 這行使用 plt.plot() 函式繪製折線圖。在這裡，data 列表中的數字將被視為 y 軸的值，
# 而 x 軸的值，因為未提供 x 軸的數值，所以它將自動以索引值 (0, 1, 2, 3, 4) 作為 x 軸。

# Line22 說明，plt.show()
# 此行顯示繪製的折線圖。 plt.show() 函式將所有的繪圖操作收集起來，然後顯示在螢幕上。
# 如果沒有此行，則圖形將不會在運行程式後顯示。


import matplotlib.pyplot as plt

data = [-1, -4.3, 15, 21, 31]
plt.plot(data)  # x軸是 0,1,2,3,4
plt.show()
```

## 繪製不同線條樣式和色彩的折線圖(Ch13_1_1a.py)

```python
# 目的是使用 Matplotlib 繪製一個具有特定樣式的折線圖。

# Line22 說明，import matplotlib.pyplot as plt
# 此行導入 Matplotlib 中的 pyplot 模組，並使用 plt 作為別名，以便在後續的程式碼中調用它的功能。

# Line24 說明，data = [-1, -4.3, 15, 21, 31]
# 這行建立一個名為 data 的 Python 列表，其中包含要在折線圖中繪製的數據點。

# Line26 說明，plt.plot(data, "o--b")
# 這行使用 plt.plot() 函式繪製折線圖。第一個參數 data 是 y 軸的數據，它將用於繪製折線圖的 y 值。
# "o--b" 是用於定義線條樣式的字符串參數。在這裡，它指定以下特性：
# "o" 表示使用圓點標記數據點。
# "--" 表示線條使用虛線。
# "b" 表示線條的顏色為深藍色 (blue)。
# 因此，繪製的折線圖將使用深藍色的虛線來表示數據點。

# Line27 說明，plt.show()
# 此行顯示繪製的折線圖。 plt.show() 函式將所有的繪圖操作收集起來，然後顯示在螢幕上。
# 如果沒有此行，則圖形將不會在運行程式後顯示。


import matplotlib.pyplot as plt

data = [-1, -4.3, 15, 21, 31]
# "o--b"是表示深藍色虛線
plt.plot(data, "o--b")  # x軸是 0,1,2,3,4
plt.show()
```

## 繪製每日攝氏溫度的折線圖(Ch13_1_1b.py)

```python
# 目的是使用 Matplotlib 繪製一個折線圖，顯示一段時間內的溫度變化。

# Line22 說明，import matplotlib.pyplot as plt
# 此行導入 Matplotlib 中的 pyplot 模組，並使用 plt 作為別名，以便在後續的程式碼中調用它的功能。

# Line24 說明，days = range(0, 22, 3)
# 這行建立一個 days 的 range 物件，其中包含溫度測量的日期，從 0 開始，到 22 （不包括 22），步長為 3。
# 因此，days 包含的數字為 0, 3, 6, 9, 12, 15, 18, 21。

# Line25 說明，celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 32.6, 33.1]
# 這行建立一個名為 celsius 的 Python 列表(list)，其中包含對應每個日期的溫度數據。

# Line26 說明，plt.plot(days, celsius)
# 這行使用 plt.plot() 函式繪製折線圖。第一個參數 days 是 x 軸的數據，第二個參數 celsius 是 y 軸的數據。
# 因此，此函式將在以 days 為 x 軸、celsius 為 y 軸的座標軸上繪製折線圖。

# Line27 說明，plt.show()
# 此行顯示繪製的折線圖。 plt.show() 函式將所有的繪圖操作收集起來，然後顯示在螢幕上。
# 如果沒有此行，則圖形將不會在運行程式後顯示。


import matplotlib.pyplot as plt

days = range(0, 22, 3) # range(start, stop, step)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5, 32.6, 33.1]
plt.plot(days, celsius)
plt.show()
```

## 繪製2張垂直排列的子圖表(Ch13_4_1.py)

```python
# 目的是使用 Matplotlib 繪製兩個垂直排列的子圖表，其中包含正弦和餘弦函數的圖形。


# 導入 Matplotlib 模組，並將其別名設為 plt。
import matplotlib.pyplot as plt

# 導入 math 模組，用於計算三角函數。
import math

# 定義了 x 軸上的數據點(列表)
x = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,
    5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]

# 根據 x 中的數據計算了正弦值、餘弦值。
sinus = [math.sin(v) for v in x]
cosinus = [math.cos(v) for v in x]

# 指定了第一個子圖的位置，其中 2 表示總行數，1 表示總列數，1 表示子圖的索引。
plt.subplot(2, 1, 1)

# 在第一個子圖中繪製了正弦函數的圖形，顏色為紅色，樣式為實線，點標記為圓點。
plt.plot(x, sinus, "r-o")

# 指定了第二個子圖的位置，其中 2 表示總行數，1 表示總列數，2 表示子圖的索引。
plt.subplot(2, 1, 2)

# 在第二個子圖中繪製了餘弦函數的圖形，顏色為綠色，樣式為虛線。
plt.plot(x, cosinus, "g--")

# 顯示繪製的圖形。
plt.show()
```

## 使用軸繪製子圖表(Ch13_4_2.py)

```python
# 目的是使用 Matplotlib 在一個圖中畫出了正弦和餘弦函數的曲線。

# 導入 Matplotlib 模組，並將其別名設為 plt。
import matplotlib.pyplot as plt

# 導入 math 模組，用於計算三角函數。
import math

# 定義了 x 軸上的數據點(列表)。
x = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,
     5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]

# 使用列表生成式，對於 x 中的每個值 v，計算其對應的正弦值，並將結果加入到 sinus 列表中。
sinus = [math.sin(v) for v in x]

# 使用列表生成式，對於 x 中的每個值 v，計算其對應的餘弦值，並將結果加入到 cosinus 列表中。
cosinus = [math.cos(v) for v in x]

# 建立一個包含兩個子圖的畫布，1 表示一行，2 表示兩列，figsize=(6,4) 設置了畫布的大小為寬度 6 單位、高度 4 單位。
fig, axes = plt.subplots(1,2, figsize=(6,4))

# 在第一個子圖中繪製了正弦函數的曲線，"r-o" 表示使用紅色的實線以及圓點標記。
axes[0].plot(x, sinus, "r-o")

# 在第二個子圖中繪製了餘弦函數的曲線，"g--" 表示使用綠色的虛線。
axes[1].plot(x, cosinus, "g--")

# 顯示繪製的圖形。
plt.show()
```
