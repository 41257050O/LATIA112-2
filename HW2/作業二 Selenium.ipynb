```python
!pip install selenium
# 導入Selenium的webdriver模組，用於控制Chrome瀏覽器
from selenium import webdriver

# 從Selenium函式庫的common.by模組中導入`By`，用於指定元素的定位方式。
import pandas as pd

from selenium.webdriver.common.by import By

# 建立ChromeOptions物件，這個物件用於設定Chrome瀏覽器的選項。
chrome_options = webdriver.ChromeOptions()

# 使用add_argument()方法添加--headless參數，使Chrome以headless模式運行，即在後台運行，不顯示GUI界面。
chrome_options.add_argument('--headless')

# 使用add_argument()方法添加--no-sandbox參數，這是一個必要的參數，用於解決在某些環境下的運行問題。
chrome_options.add_argument('--no-sandbox')

# 將 headless 屬性設置為 True。
chrome_options.headless = True

# 初始化Chrome WebDriver，並傳遞Chromedriver的路徑和選項。
driver = webdriver.Chrome(options=chrome_options)

# 設置隱式等待時間為10秒。隱式等待是在查找元素時的一種等待方式，
# 如果WebDriver沒有找到元素，它會等待一定的時間再進行查找，直到超過設定的時間為止。
driver.implicitly_wait(10)
data=[]


# 使用get方法訪問指定的URL，在此是"https://www.flag.com.tw/books/school_code_n_algo"。
driver.get("https://czbooks.net/c/favorite/total")

title = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/ul/li[22]/a')
pwd1 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[1]/div/div[1]/a/div[1]/img")
bookname1 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[1]/div/div[1]/a/div[3]")
author1 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[1]/div/div[2]")
chapter1 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[1]/div/div[3]")
collected1 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[1]/div/div[5]/ul/li[1]")
watched1 = driver.find_element(By.XPATH,'/html/body/div[5]/div/ul[1]/li[1]/div/div[5]/ul/li[2]')

pwd2 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[3]/div/div[1]/a/div[1]/img")
bookname2 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[3]/div/div[1]/a/div[3]")
author2 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[3]/div/div[2]")
chapter2 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[3]/div/div[3]")
collected2 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[3]/div/div[5]/ul/li[1]")
watched2 = driver.find_element(By.XPATH,'/html/body/div[5]/div/ul[1]/li[3]/div/div[5]/ul/li[2]')

pwd3 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[2]/div/div[1]/a/div[1]/img")
bookname3 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[2]/div/div[1]/a/div[3]")
author3 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[2]/div/div[2]")
chapter3 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[2]/div/div[3]")
collected3 = driver.find_element(By.XPATH,"/html/body/div[5]/div/ul[1]/li[2]/div/div[5]/ul/li[1]")
watched3 = driver.find_element(By.XPATH,'/html/body/div[5]/div/ul[1]/li[2]/div/div[5]/ul/li[2]')

temp1={
    "name":bookname1.text,
    "author":author1.text,
    "chapter":chapter1.text,
    "收藏數":collected1.text,
    '觀看數':watched1.text,
    "img url":pwd1.get_attribute("src")

}
temp2={
    "name":bookname2.text,
    "author":author2.text,
    "chapter":chapter2.text,
    "收藏數":collected2.text,
    '觀看數':watched2.text,
    "img url":pwd2.get_attribute("src")

}

temp3={
    "name":bookname3.text,
    "author":author3.text,
    "chapter":chapter3.text,
    "收藏數":collected3.text,
    '觀看數':watched3.text,
    "img url":pwd3.get_attribute("src")

}
data.append(temp1)
data.append(temp2)
data.append(temp3)

print(title.text)
print('---------------------------')
print(bookname1.text)
print(author1.text)
print(chapter1.text)
print('收藏數'+collected1.text)
print('觀看數'+watched1.text)
print(pwd1.get_attribute("src"))
print('---------------------------')

print(bookname2.text)
print(author2.text)
print(chapter2.text)
print('收藏數'+collected2.text)
print('觀看數'+watched2.text)
print(pwd2.get_attribute("src"))

print('---------------------------')

print(bookname3.text)
print(author3.text)
print(chapter3.text)
print('收藏數'+collected3.text)
print('觀看數'+watched3.text)
print(pwd3.get_attribute("src"))


print('---------------------------')
df = pd.DataFrame(data)
df.to_csv("book1.csv", encoding='utf_8_sig', index=False)
print(df)
```
