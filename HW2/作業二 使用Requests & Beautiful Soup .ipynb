從小說狂人中抓熱門排行排名(圖片、小說名稱、作者、收藏數、觀看數)

```python
from re import findall
!pip install selenium
import pandas as pd
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

url = "https://czbooks.net/c/normal/total"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

books=soup.find_all('li', class_='novel-item-wrapper')

data = []
index = 1

for book in books:
    bookname=book.find('div', class_='novel-item-title').text
    author_link = book.find('div', class_='novel-item-author').a
    author = author_link.text
    image_url = book.find('img')['src']
    status=book.find('div',class_='novel-item-status')
    state = status.find_all('li')
    watched =state[0].text
    collected = state[1].text

    temp_data = {
          "title": bookname,
          'author': author,
          "ImageURL" : image_url,
          "Watched": watched,
          "Collected": collected
        }
    data.append(temp_data)

df = pd.DataFrame(data)
df.to_csv("book.csv", encoding='utf_8_sig', index=False)
print(df)
```
